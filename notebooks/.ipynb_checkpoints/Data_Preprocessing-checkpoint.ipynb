```python
# Import necessary libraries and load your dataset
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
current_notebook_path = os.path.abspath('__file__')
```

```python

# Navigate to the parent directory (one level up) to access "data/raw"
data_file_path = os.path.join(os.path.dirname(os.path.dirname(current_notebook_path)), 'data', 'raw', 'diabetes.csv')
```

```python
# Load the data into a Pandas DataFrame
data = pd.read_csv(data_file_path)
```

```python
# Data Preprocessing
# ------------------

# 1. Handling Missing Values
# Check for missing values
missing_values = data.isnull().sum()
```

```python
# Display the count of missing values for each column
print("Missing Values Before Handling:")
print(missing_values)
```

```python
# Fill missing values with the mean of each column
data.fillna(data.mean(), inplace=True)
```

```python

# Fill missing values with the mean of each column
data.fillna(data.mean(), inplace=True)
```

```python
# Check if missing values are completely handled
missing_values_after_fill = data.isnull().sum()
```

```python
# Display the count of missing values after filling
print("\nMissing Values After Handling:")
print(missing_values_after_fill)
```

```python
# Display the first few rows of the dataset to get an overview
print(data.head())
```

```python
# Summary statistics
print(data.describe())
```

```python
# Count the number of instances in each class (assuming a binary classification problem)
class_counts = data['Glucose'].value_counts()
print(class_counts)
```

```python
print(data.head())
# Display the first few rows of the dataset to understand its structure
print(data.head())
```

```python
# Count the number of instances in each class (assuming a binary classification problem)
class_counts = data['Outcome'].value_counts()
print(class_counts)
```

```python
# Pairplot to visualize relationships between numerical features
sns.pairplot(data, hue='Outcome', markers=['o', 's'], palette='husl')
plt.title('Pairplot of Numerical Features')
plt.show()
```

```python
# Box plots to visualize feature distributions by outcome class
plt.figure(figsize=(12, 6))
for i, column in enumerate(data.columns[:-1]):  # Exclude the 'Outcome' column
    plt.subplot(2, 4, i + 1)
    sns.boxplot(x='Outcome', y=column, data=data, palette='husl')
    plt.title(f'Boxplot of {column}')
plt.tight_layout()
plt.show()
```

```python
# Correlation heatmap
correlation_matrix = data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Histogram for Age
plt.figure(figsize=(8, 6))
sns.histplot(data['Age'], bins=20, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Boxplot for Glucose
plt.figure(figsize=(8, 6))
sns.boxplot(x='Outcome', y='Glucose', data=data)
plt.title('Glucose Boxplot by Outcome')
plt.xlabel('Outcome')
plt.ylabel('Glucose')
plt.show()
```

```python
# Class distribution
plt.figure(figsize=(6, 4))
sns.countplot(data['Outcome'])
plt.title('Class Distribution (0: No Diabetes, 1: Diabetes)')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

```python
# Boxplot for Glucose
plt.figure(figsize=(8, 6))
sns.boxplot(x='Outcome', y='Glucose', data=data)
plt.title('Glucose Boxplot by Outcome')
plt.xlabel('Outcome')
plt.ylabel('Glucose')
plt.show()
```

```python
# Countplot for Pregnancies
plt.figure(figsize=(8, 6))
sns.countplot(x='Pregnancies', hue='Outcome', data=data)
plt.title('Pregnancies Countplot')
plt.xlabel('Number of Pregnancies')
plt.ylabel('Count')
plt.show()
```
